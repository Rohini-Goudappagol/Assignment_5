3. Explain the use of grid-auto-row using code examples.

      - <!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>grid-auto-row</title>
    <!--
        Items are placed in rows or columns not explicitly defined.

        grid-auto-row:
            Defines the height of rows that are automatically created.
            We defined only 2 columns, no rows.
            When items overflow, new rows are automatically added.
            Those rows use height from grid-auto-rows: 80px.
    -->
    <style>
        .container{
            display: grid;
            grid-template-columns: 100px 100px;
            grid-auto-rows: 80px;
            border: 2px solid;
        }
        .box{
            background: lightgreen;
            border: 1px solid #000;
        }
    </style>
</head>
<body>
    <div class="container">
        <div class="box one">1</div>
        <div class="box two">2</div>
        <div class="box three">3</div>
        <div class="box four">4</div>
        <div class="box five">5</div>

    </div>
</body>

5. Explain the difference between justify-items and justify-self using code examples
    - <!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>justify-items and justify-self</title>
    <style>
        /*
        justify-items
            1. Applies to grid containers.
            2. Aligns all items in the grid horizontally within their grid area.

        justify-self
            1. Applies to individual grid items.
            2. Overrides justify-items for that specific item only.
         */
        .container{
            display: grid;
            grid-template-columns: repeat(3,1fr)  ;
            padding: 10px;
            border: 2px solid;
            justify-content: space-between;
            gap: 0.5rem;
            justify-items: stretch; /* applies to all the grid items in the container*/
            align-items: center;
            height: 300px;
            text-align: center;
           
        }
        .box{
            background-color: aqua;
            margin: 4px;
            border: 1px solid;
        }
        .two {
            justify-self: start;
        }
        .five{
            justify-self: end;
        }
        .four{
            justify-self: end;
        }
    </style>
</head>
<body>
    <div class="container">
        <div class="box one">1</div>
        <div class="box two">2</div>
        <div class="box three">3</div>
        <div class="box four">4</div>
        <div class="box five">5</div>
        <div class="box six">6</div>
        
    </div>
</body>
</html>
